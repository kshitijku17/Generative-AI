```python
#pie plot
import matplotlib.pyplot as plt
import numpy as np
fruit_percentage = np.array([20,25,35,20])
fruit_name = np.array(["Apple", "Banana", "Orange", "Watermelon"])
color = np.array(["red", "yellow", "orange", "green"])
myexplode = np.array([0.1,0.1,0.1,0.1])
plt.pie(fruit_percentage, labels=fruit_name, autopct="%1.1f%%",colors=color,explode=myexplode)
plt.legend(title = "Fruit",loc ="best")
plt.show()
```

```python
#barh plot--- It's work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100,120,90,80])
color = np.array(["red", "yellow", "orange", "green"])
plt.barh(fruit,quantity,color=color)
plt.ylabel("Fruit")
plt.xlabel("Quantity")
plt.title("Fruit vs Quantity")
plt.show()
```

```python
#bar plot--- It's work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100,120,90,80])
color = np.array(["red", "yellow", "orange", "green"])
plt.bar(fruit,quantity,color=color)
plt.xlabel("Fruit")
plt.ylabel("Quantity")
plt.title("Fruit vs Quantity")
plt.show()
```

```python
age_of_car = np.array([4,5,6,7,8,9,10,11,12,])
speed_of_car = np.array([150,140,130,120,110,100,90,80,70])
plt.scatter(age_of_car,speed_of_car, color="black")
plt.xlabel("Age")
plt.ylabel("Speed")
plt.title("Age vs Speed")
plt.show()
```
